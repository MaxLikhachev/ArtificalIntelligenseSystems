# Трек
Хранит в себе основную информацию о музыкальном треке: темп, название, автора, массив нот (мелодию) 
```
track: {
  tempo: 120, // темп мелодии в bpm
  title: "Title", // название трека 
  author: "Author", // автор трека
  notes: [note] // последовательность нот (мелодия)
}
```

# Нота
Хранит в себе информацию о ноте: ее номер в мелодии, расположение в звукоряде относительно первой ноты мелодии, длительность и отклонение (транспонирование) от оригинальной ноты.  
```
note: {
  index: 0, // номер ноты в мелодии
  semitone: 0, // расположение ноты в звукоряде в полутонах (относительно первой ноты)
  duration: 0, // длительность ноты 
  transpose: 0 // транспонирование в полутонах от оригинальной ноты
}
```
## Номер ноты в мелодии
Целое число (положительное), показывающее местоположение ноты в мелодии. В случае монофоничной мелодии является уникальным. В случае необходимости хранения интервалов и аккордов может повторяться. 

Пример (монофония):
*   Мелодия: C#4 G4 A#3 D4
*   Номера нот: 
 *   C#4: 0 
 *   G4: 1
 *   A#3: 2
 *   D4: 3


Пример (полифония):
*   Мелодия: C#4 G4 (A#3+F3+A#4) D4
*   Номера нот: 
 *   C#4: 0 
 *   G4: 1
 *   A#3: 2
 *   F3: 2
 *   A#4: 2
 *   D4: 3

## Относительное расположение ноты в звукоряде
Целое число (от -11 до 11), показывающее отклонение в полутонах от первой ноты мелодии. Данный формат необходим для приведения всех мелодий в одну музыкальную тональность и октаву. 

Пример:
*   Оригинальная мелодия: C#4 G4 A#3 D4 (1+(4 * 12) 7+(4 * 12) 11+(3 * 12) 2+(4 * 12) = 49 55 47 50)
*   Относительное расположение нот: 0 6 -2 1
 
## Длительность ноты
Число с плавающей запятой (положительное), показывающее длительность ноты в такте 4/4 (что означает, что в 1 такте 4 четвертые ноты, 8 восьмых, и т.д.). 

Примеры:
*   Двойная целая: 2.0
*   Целая: 1.0
*   Половинная: 0.5
*   Четвертая: 0.25
*   Восьмая: 0.125
*   Шестнадцатая: 0.0625
*   Тридцать вторая: 0.03125
*   Шестьдесят четвертая: 0.15625

Стандартные размерности представляют из себя последовательность степеней числа 2 согласно своему названию. В случае других размерностей (с точкой, с двумя точками, триоли, квантоли, секстоли, и т.д.) длительность ноты получается путем умножения на соответствующий коэффициент (с точкой: *1.5, и т.д.).

## Транспонирование
Целое число (как положительное, так и отрицательное), показывающее количество полутонов (разницу между исходной и полученной нотой). Необходимо для обратной конвертации в исходную ноту.

Пример:
*   Оригинальная тональность (и сама первая нота, 0 - C, 1 - C#, и т.д.): 1
*   Оригинальная октава (0 - субконтроктава): 4
*   Транспонирование: 4 * 12 + 1 = 49
